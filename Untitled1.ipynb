**Q.15. How we can use histogram check individual color components...??**

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
image =cv2.imread('color.jpg')
histogram =cv2.calcHist([image],[0],None,[256],[0,256])
# we plot histogram, ravel() flatens our image array
plt.hist(image.ravel(),256,[0,256]);
plt.show()

color =('b','g','r') # viewing separate color channel
for i, col in enumerate(color):
    histogram2 =cv2.calcHist([image],[i],None,[256],[0,256])
    plt.plot(histogram2,color=col)
    plt.xlim([0,256])
plt.show()
    
```

**Q.21. What iS THE AFFINE TRANSFORMATION ...?**

```python
import cv2
import numpy as np
image = cv2.imread('color.jpg')
# store height and width of image
height,width = image.shape[:2]
quarter_height,quater_width = height/4,width/4
# T = : 1  0  Tx : 
#     : 0  1  Ty :
T = np.float32([[1,0,quater_width],[0,1,quarter_height]])

# here we Affine to trasfrom the image using the matrx
img_translation = cv2.warpAffine(image,T,(width,height))
cv2.imshow('Translation',img_translation)
cv2.waitKey()
cv2.destroyAllWindows()
```

**Q.22. How we rotate the image ...?**

```python
import cv2
import numpy as np
image = cv2.imread('dog.jpg')

height,width = image.shape[:2]
# divide by 2 to rotation the image around its centre

rotation_matrix =cv2.getRotationMatrix2D((width/2,height/2),270,.5)
# x_centre,Y_centre,angle,scale

# here we Affine to trasfrom the image using the matrx
img_translation = cv2.warpAffine(image,rotation_matrix,(width,height))
cv2.imshow('Rotated Image',img_translation)
cv2.imshow('Original Image',image)
cv2.waitKey()
cv2.destroyAllWindows()
```

**Another Method**

```python
import cv2
import numpy as np
image = cv2.imread('dog.jpg')
img_translation = cv2.transpose(image)
cv2.imshow('Rotated Image',img_translation)
cv2.imshow('Original Image',image)
cv2.waitKey()
cv2.destroyAllWindows()
```
